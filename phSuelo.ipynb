```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ejercicioUno - Hoja 1.csv')
```

```python
df
```

```python
dfAcido = df[df['ph'] < 5.8]
```

```python
dfAcido
```

```python

dfBasico = df[df['ph'] > 6.8]
```

```python
dfNeutro = df[(df['ph'] >= 5.8) & (df['ph'] <= 6.8)]
```

```python
dfNeutro
```

```python
plt.hist(dfBasico['ph'], bins=10)
plt.xlabel("pH")
plt.ylabel("Frecuencia")
plt.title("Histograma de pH (b√°sico)")
plt.show()
```

```python
plt.hist(dfNeutro['ph'], bins=10)
plt.xlabel("pH")
plt.ylabel("Frecuencia")
plt.title("Histograma de pH (Neutro)")
plt.show()
```

```python
plt.hist(dfAcido['ph'], bins=10)
plt.xlabel("pH")
plt.ylabel("Frecuencia")
plt.title("Histograma de pH (Acido)")
plt.show()
```

```python
plt.hist(df['ph'], bins=10)
plt.xlabel("pH")
plt.ylabel("Frecuencia")
plt.title("Histograma de pH (Totales)")
plt.show()
```

```python
plt.scatter(df['muestra_id'], df['ph'])
plt.xlabel("Muestra")
plt.ylabel("pH")
plt.title("pH por muestra")
plt.show()
```
